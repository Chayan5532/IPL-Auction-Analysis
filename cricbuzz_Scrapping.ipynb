```python
# URL
url_25="https://www.cricbuzz.com/cricket-series/ipl-2025/auction/completed"

# Driver
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
driver.get(url_25)

# Initial delay (important)
time.sleep(5)

# -------------------- SCROLLING LOGIC --------------------

last_height = driver.execute_script("return document.body.scrollHeight")

for scroll in range(20):  # adjust if more rows load
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(3)  # delay to avoid blocking
    
    new_height = driver.execute_script("return document.body.scrollHeight")
    
    if new_height == last_height:
        break
    
    last_height = new_height

# -------------------- PARSING --------------------

soup = BeautifulSoup(driver.page_source, 'html.parser')

features = soup.find_all(
    'div',
    class_="grid grid-cols-[4.1fr,2fr,1.3fr,1.3fr,1.3fr]"
)

print(f"Total rows scraped: {len(features)}")

# -------------------- COLUMN LISTS --------------------

player = []
role = []
country = []
status = []
base_price = []
final_price = []
team = []

# -------------------- EXTRACTION LOOP --------------------

for i in features:
    a = i.text.strip()

    # -------- PLAYER (PASTE YOUR WORKING REGEX HERE) --------
    p = re.findall(r'^([A-Za-z\s]+?)(Bowler|Batter|Allrounder|WK-Batter)', a)
    player.append(p[0][0].strip() if p else np.nan)

    # -------- ROLE --------
    r = re.findall(r'(Bowler|Batter|Allrounder|WK-Batter)', a)
    role.append(r[0] if r else np.nan)

    # -------- COUNTRY --------
    c = re.findall(r'(Bowler|Batter|Allrounder|WK-Batter)([A-Za-z\s]+?)(sold|unsold)', a)
    country.append(c[0][1].strip() if c else np.nan)

    # -------- STATUS --------
    s = re.findall(r'(sold|unsold)', a)
    status.append(s[0] if s else np.nan)

    # -------- BASE PRICE --------
    bp = re.findall(r'(sold|unsold)(\d+(?:\.\d+)?\s(?:L|Cr))', a)
    base_price.append(bp[0][1] if bp else np.nan)

    # -------- FINAL PRICE --------
    fp = re.findall(r'(?:L|Cr)(\d+(?:\.\d+)?\s(?:L|Cr)|--)', a)
    final_price.append(fp[0] if fp else '--')

    # -------- TEAM (FIXED LOGIC) --------
    t = re.findall(r'([A-Z]{2,3})$', a)
    team.append(t[0] if t else '--')

    # Small delay per row (safe scraping)
    time.sleep(0.05)

# -------------------- FINAL DATAFRAME --------------------

df = pd.DataFrame({
    'Player': player,
    'Role': role,
    'Country': country,
    'Status': status,
    'Base Price': base_price,
    'Final Price': final_price,
    'Team': team
})

driver.quit()

df
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import pandas as pd
import time

def scrape_ipl_auction(year):
    url = f"https://www.cricbuzz.com/cricket-series/ipl-{year}/auction/completed"
    driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
    driver.get(url)
    time.sleep(5)

    #  Scroll dynamically until no new players load
    last_height = driver.execute_script("return document.body.scrollHeight")
    while True:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(2)  # wait for loading
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:  # no new content
            break
        last_height = new_height

    html = driver.page_source
    soup = BeautifulSoup(html, "html.parser")

    players = []
    cards = soup.select(f"a[href^='/cricket-series/ipl-{year}/auction/players/']")
    print(f"Year {year}  Cards found: {len(cards)}")

    for card in cards:
        try:
            name = card.select_one("div.font-bold").get_text(strip=True)
            role_country = card.select("p")
            role = role_country[0].get_text(strip=True) if len(role_country) > 0 else ""
            country = role_country[1].get_text(strip=True) if len(role_country) > 1 else ""
            
            base_price = card.find("span", string="Base Price")
            base_price = base_price.find_next("span").get_text(strip=True) if base_price else ""
            
            final_price = card.find("span", string="Final Price")
            final_price = final_price.find_next("span").get_text(strip=True) if final_price else ""
            
            team = card.find("span", string="Team")
            team = team.find_next("span").get_text(strip=True) if team else ""
            
            status = card.select_one("div.uppercase.text-white")
            status = status.get_text(strip=True) if status else ""

            players.append({
                "Player": name,
                "Role": role,
                "Country": country,
                "Status": status,
                "Base Price": base_price,
                "Final Price": final_price,
                "Team": team
            })
        except Exception:
            continue

    driver.quit()
    return pd.DataFrame(players)

# Scrape all 3 years
all_years = [2024, 2025, 2026]
df_list = []

for year in all_years:
    df_year = scrape_ipl_auction(year)
    df_year['Year'] = year
    df_list.append(df_year)

final_df = pd.concat(df_list, ignore_index=True)

# Display final data
pd.set_option('display.max_rows', None)
print(final_df)
print("Total players scraped:", len(final_df))
```
